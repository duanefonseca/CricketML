```python
import pandas as pd
```

```python
# prompt: load file /content/dados_egressos_ICB.csv

df = pd.read_csv('/content/dados_egressos_ICB.csv', encoding='latin-1')
df.head()
```

```python
# prompt: show df column

print(df.columns)
```

```python
# prompt: type of the columns

print(df.dtypes)
```

```python
# prompt: transform from float to int64 columns ['ano_inicio_ativ_atual', 'no_inicio_ativ_ant', 'ano_fim_ativ_ant ', 'pos_ing', 'pos_conc ']

for column in ['ano_inicio_ativ_atual', 'ano_inicio_ativ_ant', 'ano_fim_ativ_ant', 'pos_ing', 'pos_conc']:
  df[column] = pd.to_numeric(df[column], errors='coerce').fillna(0).astype('Int64')

print(df.dtypes)
```

```python
df.head()
```

```python
# prompt: show counts of ['Autonomo_diferente' 'Sem_trabalho' 'Publico_area' 'CLT_area'
#  'CLT_diferente' 'Autonomo_area'] in 'classe_ativ_atual'

counts = df['classe_ativ_atual'].value_counts()
print(counts)
```

```python
# prompt: consider only 'Publico_area' in 'classe_ativ_atual'. Calculate the average of the diference between 'ano_inicio_ativ_atual' and 'ano_conc'
# print mean +- SE
# plot it as mean +- SE

import numpy as np
import matplotlib.pyplot as plt

df_publico = df[df['classe_ativ_atual'] == 'Publico_area']

diferenca_anos = df_publico['ano_inicio_ativ_atual'] - df_publico['ano_conc']

media = np.mean(diferenca_anos)
erro_padrao = np.std(diferenca_anos) / np.sqrt(len(diferenca_anos))

print(f"Média da diferença entre ano_inicio_ativ_atual e ano_conc: {media:.2f}")
print(f"Erro Padrão: {erro_padrao:.2f}")

plt.errorbar(1, media, yerr=erro_padrao, fmt='o')
plt.xlabel("Publico_area")
plt.ylabel("Diferença entre ano_inicio_ativ_atual e ano_conc")
plt.title("Média da Diferença entre ano_inicio_ativ_atual e ano_conc (Publico_area)")
plt.show()
```

```python
# prompt:  consider only 'Publico_area' in 'classe_ativ_atual'. Calculate the average of the diference between 'ano_inicio_ativ_atual' and TODAY

from datetime import date

df_publico = df[df['classe_ativ_atual'] == 'Publico_area']

hoje = date.today().year

diferenca_anos = hoje - df_publico['ano_inicio_ativ_atual']

media = np.mean(diferenca_anos)
erro_padrao = np.std(diferenca_anos) / np.sqrt(len(diferenca_anos))

print(f"Média da diferença entre ano_inicio_ativ_atual e hoje: {media:.2f}")
print(f"Erro Padrão: {erro_padrao:.2f}")

plt.errorbar(1, media, yerr=erro_padrao, fmt='o')
plt.xlabel("Publico_area")
plt.ylabel("Diferença entre ano_inicio_ativ_atual e hoje")
plt.title("Média da Diferença entre ano_inicio_ativ_atual e hoje (Publico_area)")
plt.show()
```

```python
# prompt: calculate average +- SE of 'nota_prof' and 'nota_pos' by each 'Sem_trabalho', 'Publico_area', 'CLT_area', 'CLT_diferente',        'Autonomo_diferente' and 'Autonomo_area'

import numpy as np

# Group by 'classe_ativ_atual' and calculate the mean and standard error for 'nota_prof' and 'nota_pos'
def calculate_mean_se(group):
  mean_nota_prof = np.mean(group['nota_prof'])
  se_nota_prof = np.std(group['nota_prof']) / np.sqrt(len(group['nota_prof'])) if len(group['nota_prof']) > 0 else np.nan
  mean_nota_pos = np.mean(group['nota_pos'])
  se_nota_pos = np.std(group['nota_pos']) / np.sqrt(len(group['nota_pos'])) if len(group['nota_pos']) > 0 else np.nan
  return pd.Series({'mean_nota_prof': mean_nota_prof, 'se_nota_prof': se_nota_prof,
                    'mean_nota_pos': mean_nota_pos, 'se_nota_pos': se_nota_pos})

result = df.groupby('classe_ativ_atual').apply(calculate_mean_se)

print(result)
```
