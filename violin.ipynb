```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_csv('/content/drive/MyDrive/arquivos_dados_colab/anova_one_way/anova.csv')

# Separar os dados com base na coluna de grupos categ√≥ricos
dados_grupo_categorical = df['B']
dados_observados = df['A']

# Create the violin plot with inner boxplot
plt.figure(figsize=(10, 6))
sns.violinplot(x=dados_grupo_categorical, y=dados_observados, data=df, inner="quartile", palette='pastel')
sns.swarmplot(x=dados_grupo_categorical, y=dados_observados, data=df, color='k', size=3)
#sns.boxplot(x=dados_grupo_categorical, y=dados_observados, data=df, boxprops=dict(facecolor='none'), width=0.3)
# Adjust y-axis limits
max_value = dados_observados.max()
plt.ylim(0, max_value + 0.3 * max_value)

# Adjust x-axis limits to add offsets at both ends
grupo = df[dados_grupo_categorical.name].unique()
if "Grupo A" in grupo and "Grupo E" in grupo:
    plt.xlim(-0.5, len(grupo) - 0.5)

# Adding the mean points
means = df.groupby(dados_grupo_categorical.name)[dados_observados.name].mean()
grupo = df[dados_grupo_categorical.name].unique()

for i, grupo in enumerate(grupo):
    plt.plot(i, means[grupo], 'ro')


# Save the plot
#plt.savefig('your_plot.png')

# Additional annotations for the statistical results can be added as needed

# Display
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the font size
font_size_axes = 20
font_size_ticks = 12
font_size_legend = 16

# Apply the font size choices
plt.rcParams['axes.labelsize'] = font_size_axes
plt.rcParams['xtick.labelsize'] = font_size_ticks
plt.rcParams['ytick.labelsize'] = font_size_ticks
plt.rcParams['legend.fontsize'] = font_size_legend

# Carregar os dados do arquivo CSV
#df = pd.read_csv('/content/anova.csv')

# Separar os dados com base na coluna de grupos categ√≥ricos
dados_grupo_categorical = df['B']
dados_observados = df['A']

# Create the violin plot with inner boxplot
plt.figure(figsize=(10, 6))
sns.violinplot(x=dados_grupo_categorical, y=dados_observados, data=df, inner="quartile", palette='pastel')
sns.swarmplot(x=dados_grupo_categorical, y=dados_observados, data=df, color='k', size=3)

# Setting the axes titles
plt.xlabel('Group Category')
plt.ylabel('Observed Values')

# Adjust y-axis limits
max_value = dados_observados.max()
plt.ylim(0, max_value + 0.3 * max_value)

# Adjust x-axis limits to add offsets at both ends
grupo = df[dados_grupo_categorical.name].unique()
if "Grupo A" in grupo and "Grupo E" in grupo:
    plt.xlim(-0.5, len(grupo) - 0.5)

# Adding the mean points
means = df.groupby(dados_grupo_categorical.name)[dados_observados.name].mean()
grupo = df[dados_grupo_categorical.name].unique()

for i, grupo in enumerate(grupo):
    plt.plot(i, means[grupo], 'ro')

# Save the plot
#plt.savefig('your_plot.png')

# Display
plt.tight_layout()
plt.show()

```

```python
import plotly.express as px
violin = px.violin(df, y="A", x="B", box=True, points="all")

violin.show()

```

```python

# Assuming 'df' is your DataFrame and it has columns 'A' and 'B'
violin = px.violin(df, y="A", x="B", box=True, points="all")

# Update the violin and box plot color and make it semi-transparent
violin.update_traces(fillcolor='rgba(255, 0, 0, 0.5)', line_color='red', opacity=0.6)

# Update the points color to match the violin and make them semi-transparent
violin.update_traces(marker=dict(color='rgba(255, 0, 0, 0.5)'), selector=dict(type='violin'))

violin.show()
```

```python
box_plot = px.box(df, y="A", x="B", points="all")
box_plot.update_traces(marker_color='red', line_color='red')
#Add mean line to the box plot # Use 'boxmean=True' to show the mean without the standard deviation
box_plot.update_traces(boxmean=True)
#Export to static HTML
box_plot.write_html("box_plot.html")
box_plot.show()
```
