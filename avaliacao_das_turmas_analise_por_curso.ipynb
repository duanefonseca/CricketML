### Baixar planilha CSV do Sistema. Antes de importar aqui, filtrar CSV para excluir registros sem resposta de ADD e para incluir somente disciplinas do QSL do curso (desmarcar QSL 'blank") e formatar as células de campos numéricos para números

```python
import pandas as pd

# Load the CSV file into a DataFrame
#avaliacao_das_turmas_analise_por_curso = pd.read_csv('/content/avaliacao_das_turmas_analise_por_curso_ADD_2022.csv')
avaliacao_das_turmas_analise_por_curso = pd.read_csv('/content/avaliacao_turmas_analise_curso_ADD_2024.csv', sep=",")

# Display the first few rows of the dataframe
avaliacao_das_turmas_analise_por_curso
```

```python
# prompt: show unique 'Nome da Disciplina'

print(avaliacao_das_turmas_analise_por_curso['Nome da Disciplina'].unique())
```

```python
# prompt: select only  'Seminário Integrador em Ciências e Biologia I'
#  'Seminário Integrador em Ciências e Biologia II'
#  'Seminário Integrador em Ciências e Biologia III' and make 'SICB' df

seminarios = [
    'Seminário Integrador em Ciências e Biologia I',
    'Seminário Integrador em Ciências e Biologia II',
    'Seminário Integrador em Ciências e Biologia III'
]
SICB_df = avaliacao_das_turmas_analise_por_curso[avaliacao_das_turmas_analise_por_curso['Nome da Disciplina'].isin(seminarios)].copy()
SICB_df
```

```python
# prompt: calculate average 'Nota ADD' by 'Nome da Disciplina' in SICB_df and plot bargraph. Plot in the same graph all data used calculate average of each bar

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average 'Nota ADD' for each 'Nome da Disciplina' in SICB_df
average_nota_add = SICB_df.groupby('Nome da Disciplina')['Nota ADD'].mean().reset_index()

# Plot the data
plt.figure(figsize=(12, 7))

# Plot the individual data points as scatter plot
sns.stripplot(x='Nome da Disciplina', y='Nota ADD', data=SICB_df, color='black', alpha=0.6, jitter=0.1)

# Plot the average 'Nota ADD' as bars
sns.barplot(x='Nome da Disciplina', y='Nota ADD', data=average_nota_add, color='skyblue')

plt.title('Average Nota ADD by Nome da Disciplina with Individual Data Points')
plt.xlabel('Nome da Disciplina')
plt.ylabel('Nota ADD')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: calculate average 'Nota ADD' by 'Nome da Disciplina' in SICB_df and plot bargraph using Plotly. Plot in the same graph all data used calculate average of each bar

import plotly.graph_objects as go
import plotly.express as px

# Calculate the average 'Nota ADD' for each 'Nome da Disciplina'
average_nota_add = SICB_df.groupby('Nome da Disciplina')['Nota ADD'].mean().reset_index()

# Create a bar trace for the average
bar_trace = go.Bar(
    x=average_nota_add['Nome da Disciplina'],
    y=average_nota_add['Nota ADD'],
    name='Average Nota ADD',
    marker_color='skyblue'
)

# Create a scatter trace for the individual data points
scatter_trace = go.Scattergl(
    x=SICB_df['Nome da Disciplina'],
    y=SICB_df['Nota ADD'],
    mode='markers',
    name='Individual Data Points',
    marker=dict(color='black', opacity=0.6, size=5)
)

# Combine the traces
fig = go.Figure(data=[bar_trace, scatter_trace])

# Update layout for better readability
fig.update_layout(
    title='Average Nota ADD by Nome da Disciplina with Individual Data Points',
    xaxis_title='Nome da Disciplina',
    yaxis_title='Nota ADD',
    xaxis=dict(tickangle=0),
    barmode='overlay' # Overlay bars and scatter points
)

# Show the plot
fig.show()
```

### Filtragem do dataframe. Somente manter registros que 'Alunos' = 'Alunos Curso'. Alternativamente, código comentado pode ser usado (<=2.5 significa 40% de Alunos Curso)

```python
# Filtering the DataFrame to keep only rows where 'Alunos' is equal to 'Alunos Curso'
#avaliacao_das_turmas_analise_por_curso_filtered = avaliacao_das_turmas_analise_por_curso[
#    avaliacao_das_turmas_analise_por_curso['Alunos'] == avaliacao_das_turmas_analise_por_curso['Alunos Curso']
#]

# Filter the dataframe to keep only the rows where the condition 'Alunos' divided by 'Alunos Curso' <= 1.25 is True
avaliacao_das_turmas_analise_por_curso_filtered = avaliacao_das_turmas_analise_por_curso[
    (avaliacao_das_turmas_analise_por_curso['Alunos'] / avaliacao_das_turmas_analise_por_curso['Alunos Curso']) <= 2.5
]

# Display the first few rows of the filtered dataframe
avaliacao_das_turmas_analise_por_curso_filtered
```

```python
avaliacao_das_turmas_analise_por_curso_filtered.columns
```

```python
# prompt: calculate mean +- ep of:
#  'Questão 1',
#        'Questão 2', 'Questão 3', 'Questão 4', 'Questão 5', 'Questão 6',
#        'Questão 7', 'Questão 8', 'Questão 9', 'Questão 10', 'Questão 11'
# Wheh NaN is found ignore it

import pandas as pd
import numpy as np
from scipy import stats

# Define the list of question columns
question_cols = [
    'Questão 1', 'Questão 2', 'Questão 3', 'Questão 4', 'Questão 5',
    'Questão 6', 'Questão 7', 'Questão 8', 'Questão 9', 'Questão 10',
    'Questão 11'
]

# Calculate the mean and standard error of the mean for each question column, ignoring NaN values
for col in question_cols:
    # Convert the column to numeric, coercing errors to NaN
    numeric_col = pd.to_numeric(avaliacao_das_turmas_analise_por_curso_filtered[col], errors='coerce')

    # Drop NaN values
    cleaned_data = numeric_col.dropna()

    if not cleaned_data.empty:
        mean_val = np.mean(cleaned_data)
        # Calculate standard error of the mean (SEM)
        sem_val = stats.sem(cleaned_data)
        print(f'{col}: {mean_val:.2f} ± {sem_val:.2f}')
    else:
        print(f'{col}: No valid data to calculate mean and SEM.')
```

```python
# prompt: calculate the average and SE  'Nota ADD' by 'Sem. QSL'

import pandas as pd
# Calculate the mean and standard error of 'Nota ADD' grouped by 'Sem. QSL'
# We need to use the filtered dataframe for this analysis
grouped_data = avaliacao_das_turmas_analise_por_curso_filtered.groupby('Sem. QSL')['Nota ADD']

average_add = grouped_data.mean()
se_add = grouped_data.sem() # .sem() calculates the standard error of the mean

# Combine the results into a new DataFrame for better presentation
summary_df = pd.DataFrame({
    'Average Nota ADD': average_add,
    'SE Nota ADD': se_add
})

summary_df
```

```python
# prompt: calculate average 'Nota ADD' by 'Nome do Professor' and save it as csv

# Calculate the average 'Nota ADD' grouped by 'Nome do Professor'
average_nota_add_by_professor = avaliacao_das_turmas_analise_por_curso_filtered.groupby('Nome do Professor')['Nota ADD'].mean()

# Convert the result to a DataFrame
average_nota_add_by_professor_df = average_nota_add_by_professor.reset_index()
average_nota_add_by_professor_df.columns = ['Nome do Professor', 'Average Nota ADD']

# Save the result as a CSV file
average_nota_add_by_professor_df.to_csv('Media_ADD_professor.csv', index=False)
average_nota_add_by_professor_df
```

```python
# prompt: order  avaliacao_das_turmas_analise_por_curso_filtered in decrescent order by  'Nota ADD' and keep only columns 'Nome da Disciplina', 'Turma', 'Nota ADD'

# Sort the filtered dataframe in descending order by 'Nota ADD'
avaliacao_das_turmas_analise_por_curso_filtered_sorted = avaliacao_das_turmas_analise_por_curso_filtered.sort_values(by=['Nota ADD'], ascending=False)

# Select only the desired columns
avaliacao_das_turmas_analise_por_curso_filtered_sorted = avaliacao_das_turmas_analise_por_curso_filtered_sorted[['Nome da Disciplina', 'Turma', 'Nota ADD']]

# Display the sorted dataframe
avaliacao_das_turmas_analise_por_curso_filtered_sorted
```

```python
# prompt: download df 'avaliacao_das_turmas_analise_por_curso_filtered_sorted' as csv

avaliacao_das_turmas_analise_por_curso_filtered_sorted.to_csv('avaliacao_das_turmas_analise_por_curso_filtered_sorted.csv', index=False)
```

### Média da ADD por disciplina, por turma

```python
import matplotlib.pyplot as plt

# Assuming 'avaliacao_das_turmas_analise_por_curso_filtered' is a predefined DataFrame

# User's choice for 'Sem. QSL', can be set to any integer from 1 to 8
escolher_semestre = 8  # Example value

# Filter the DataFrame where 'Sem. QSL' equals the user's choice
avaliacao_das_turmas_analise_por_curso_filtered_semestre = avaliacao_das_turmas_analise_por_curso_filtered[
    avaliacao_das_turmas_analise_por_curso_filtered['Sem. QSL'] == escolher_semestre
]

# Plotting the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size as desired
bars = plt.bar(avaliacao_das_turmas_analise_por_curso_filtered_semestre['Nome da Disciplina'], avaliacao_das_turmas_analise_por_curso_filtered_semestre['Nota ADD'])
plt.ylabel('Nota média ADD, por turma')  # Y-axis label
plt.title('Média da nota ADD por disciplina - Semestre: {}'.format(escolher_semestre))  # Title of the graph

# Adding the text on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

# Rotate x-axis labels to 45 degrees and align horizontally to avoid overlap
plt.xticks(rotation=15, ha='right')

plt.tight_layout()  # Adjust layout to prevent clipping of ylabel/ticklabels
plt.show()  # Display the plot
```
